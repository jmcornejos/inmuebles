```python
%sql
-- Tabla en bronze
DROP TABLE IF EXISTS bronze.brz_inmuebles;
CREATE TABLE bronze.brz_inmuebles (
    Referencia LONG,
    FechaAlta String,
    Tipo String,
    Operacion String,
    Ciudad String,
    Superficie LONG,
    Precio_Venta LONG,
    Fecha_Venta String,
    Vendedor String,
    Estatus String,
    Dias_para_Vender LONG,
    Fecha_Proceso TIMESTAMP
);
```
