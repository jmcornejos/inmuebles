```python
import pandas as pd
import numpy as np
from pyspark.sql.functions import current_timestamp
```

```python
try:
    table_brz = "bronze.brz_inmuebles"

    pdf = pd.read_csv("/Volumes/workspace/ingest/data/inmuebles.csv", sep=',', header=0)

    df = spark.createDataFrame(pdf)

    # Agregar columna fecha_proceso
    df = df.withColumn("Fecha_Proceso", current_timestamp())

    # Rename columns to match existing table schema
    df = df.withColumnRenamed("Operación", "Operacion") \
        .withColumnRenamed("Precio Venta", "Precio_Venta") \
        .withColumnRenamed("Fecha Venta", "Fecha_Venta") \
        .withColumnRenamed("Días para Vender", "Dias_para_Vender")


    # Verificar si la tabla existe
    if not spark.catalog.tableExists("bronze.brz_inmuebles"):
        # Crear tabla Delta y escribir datos
        df.write.format("delta") \
            .mode("overwrite") \
            .saveAsTable("bronze.brz_inmuebles")
    else:
        # Insertar datos en la tabla existente
        df.write.format("delta") \
            .mode("overwrite") \
            .saveAsTable("bronze.brz_inmuebles")
 
except Exception as e:
    print(f"Tipo de error: {type(e).__name__}, Mensaje: {e}")
```
